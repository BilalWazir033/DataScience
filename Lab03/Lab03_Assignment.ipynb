### Part A: Bernoulli Distribution 

```python
# probability of success
success=sum(win_data)
total=len(win_data)
p=success/total
print("Probability of Success: ",p)
```

```python
# Probability of failure
q=1-p
print("Probability of Failure: ",q)
```

```python
# calculate mean and variance
mean=p
var=p*q
print("Mean: ",mean)
print("Variance: ",var)
```

```python
# Bernoulli Probability Mass Function
print("Probability Mass Function, PMF")
print("P(X=1)=",p)
print("P(X=1)=",q)
```

```python
import matplotlib.pyplot as plt

labels=["Fail(0)", "Pass(1)"]
colors=["red","green"]
plt.bar(labels,[q,p],color=colors)
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.title("Bernoulli Distribution - Success vs Failure")

plt.show()
```

### Interpretation

- The binary variable "Success" was modeled using a Bernoulli distribution.
- The probability of success (p) represents the proportion of companies with positive profit.
- The probability of failure (1 − p) represents companies with zero or negative profit.
- The mean of the Bernoulli distribution equals p, indicating the average success rate.
- The variance p(1−p) measures variability in company success outcomes.
- From the results, if p is high, most companies are profitable; if p is low, many companies are experiencing losses.

##  Comparison Between Bernoulli and Binomial Distribution

###  Bernoulli Distribution

- Only **1 trial**
- Only **two possible outcomes**: 0 (Failure) or 1 (Success)
- Mean:  
  \[
  E(X) = p
  \]
- Variance:  
  \[
  Var(X) = p(1 - p)
  \]

---

###  Binomial Distribution

- Consists of **multiple independent trials (n trials)**
- Counts the **total number of successes** in n trials
- Mean:  
  \[
  E(X) = np
  \]
- Variance:  
  \[
  Var(X) = np(1 - p)
  \]

---

##  Interpretation

The **Bernoulli distribution** models the success or failure of a single company.

The **Binomial distribution** models the number of successful companies out of **n selected companies**.

As the number of trials \( n \) increases, the Binomial distribution becomes more spread out and gradually approaches a Normal distribution shape (according to the Central Limit Theorem).
